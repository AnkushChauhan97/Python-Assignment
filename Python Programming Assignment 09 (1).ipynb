**Q1.** Write a Python program to check if the given number is a Disarium Number?

```python
def disnum(num):
  num= str(num)
  dis= 0
  for it in range(1,len(num)+1):
    dis= dis +(int(num[it-1])**it)
  if dis== int(num):
    print('{} is a Disarium number.'.format(num))
  else:
    print('{} is not a Disarium number.'.format(num))

disnum(175)
disnum(1555)
disnum(10)
```

**Q2.** Write a Python program to print all disarium numbers between 1 to 100?

```python
def dislist(a,b):
  dislist= []
  for it in range(a,b+1):
    num= str(it)
    dis= 0
    for j in range(1,len(num)+1):
      dis= dis +(int(num[j-1])**j)
    if dis== int(num):
      dislist.append(dis)
  print(dislist)

dislist(1,100)
```

**Q5.** Write a Python program to determine whether the given number is a Harshad Number?

```python
def harnum():
    num = input('Enter a Number: ')
    sum = 0
    for it in range(len(num)):
        sum = sum + int(num[it])
    if int(num)%sum == 0:
        print(f'{num} is a Harshad Number')
    else:
        print(f'{num} is a Not Harshad Number')
        
harnum()
```

**Q6.** Write a Python program to print all pronic numbers between 1 and 100?

```python
def pronum(start,end):
    out = []
    for it in range(start,end+1):
        out.append((it)*(it+1))
        if out[-1]>=90:
          break
    print(out)
                   
pronum(1,100)
```
