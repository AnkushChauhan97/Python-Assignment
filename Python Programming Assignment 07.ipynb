**Q1.** Write a Python Program to find sum of array?

```python
array= [1,59,3,5,45,7,8]
sum= 0
for it in range(0,len(array)):
  sum= sum + array[it]

print('Sum of array= ', sum)
```

**Q5.** Write a Python Program to check if given array is Monotonic?

```python
def mono(*args):
  x= [*args]
  print(x)
  # x= eval(input('Enter the array.'))
  if all(x[i]<=x[i+1] for i in range(len(x)-1)):
    print('The array is Monotonical Increasing.')
  elif all(x[i]>=x[i+1] for i in range(len(x)-1)):
    print('The array is Monotonical Decreasing.')
  else:
    print('The aray is not Monotonic.')

mono(1,2,3,4,5,6)
mono(8,5,2,1)
mono(5,8,48,12)
```
