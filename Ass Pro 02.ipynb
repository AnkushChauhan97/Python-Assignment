**Q1.** Write a Python program to convert kilometers to miles?

```python
km= int(input('Enter value in Kilometers:'))
ml= km*0.62
print('{} km is equals to {} mile'.format(km, ml))
```

**Q2.** Write a Python program to convert Celsius to Fahrenheit?

```python
cl= int(input('Enter value in Celsius:'))
fr= (cl*(9/5))+32
print('{} Celsius is equals to {} Fahrenheits'.format(cl, fr))
```

**Q4.** Write a Python program to solve quadratic equation?

```python
import cmath
import math

a= int(input('Enter the value pof a: '))
b= int(input('Enter the value pof b: '))
c= int(input('Enter the value pof c: '))

disc = b*b-4*a*c

if disc== 0:
    r1 = -b/2*a
    r2 = -b/2*a
    print("Roots are Real",r1,r2)
elif disc > 0:
    r1 = (-b-math.sqrt(disc))/(2 * a)
    r2 = (-b+math.sqrt(disc))/(2 * a)
    print("Roots are Real and different",r1,r2)
else:
    r1 = (-b-cmath.sqrt(disc))/(2 * a)
    r2 = (-b+cmath.sqrt(disc))/(2 * a)
    print("Roots are Imaginary",r1,r2)
```

**Q5.** Write a Python program to swap two variables without temp variable?

```python
num1 = int(input('Enter first number: '))
num2 = int(input('Enter second number: '))

num1 = num1+num2
num2 = num1-num2
num1 = num1-num2
print('Numbers are swaped: First= {}, Second= {}'.format(num1,num2))
```
