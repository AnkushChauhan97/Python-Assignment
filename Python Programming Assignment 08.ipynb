**Q1.** Write a Python Program to Add Two Matrices?

```python
def addmatric(a,b):
    print('Inputs: {},{}'.format(a,b))
    if len(a) == len(b):
        out = []
        for it in range(len(a)):
            if len(a[it]) == len(b[it]):
                out.append([])
                for sit in range(len(a[it])):
                    out[it].append(a[it][sit]+b[it][sit])
            else:
                print('No of rows and columns of both matrices should same') 
    else:
        print('No of rows and columns of both matrices should same')
    print('Sum: {}'.format(out))

addmatric([[1,5,9,],[8,1,4],[6,3,8]],[[4,5,8],[8,9,3],[7,6,2]])
```

**Q2.** Write a Python Program to Multiply Two Matrices?

```python
a= [[2,5,3],[1,5,7]]
b= [[5,6],[2,8],[6,4]]

def matmulti (a,b):
  out= []
  if len(a[0])==len(b):
    for x in range(len(a)):
      out.append([0 for y in range(len(b[0]))])
    for i in range(len(a)):
      for j in range(len(b[0])):
        for k in range(len(b)):
          out[i][j] +=(a[i][k]*b[k][j])
    print(out)     
  else:
    print('Number of column of first matric should be equal to number of rows in second matric')

matmulti(a,b)
```

**Q3.** Write a Python Program to Transpose a Matrix?

```python
a= [[5,6],[2,8],[6,4]]

def trans(a):
  out= []
  for x in range(len(a[0])):
    out.append([0 for y in range(len(a))])
  for i in range(len(a[0])):
    for j in range(len(a)):
      out[i][j] =(a[j][i])
  print(out) 

trans(a)
```

**Q4.** Write a Python Program to Sort Words in Alphabetic Order?

```python
string = input("Enter a sentence: ")
sort = sorted(string.split(' '))
print(' '.join(sort))
```

**Q5.** Write a Python Program to Remove Punctuation From a String?

```python
import string

strn= input('Please enter a string with punctuations: ')

out = ''
for it in strn:
  if it not in string.punctuation:
    out += it
    
print(out)
```
